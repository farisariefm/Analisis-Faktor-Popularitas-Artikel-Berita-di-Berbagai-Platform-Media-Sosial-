**4.7 Outliers Analysis**

```python
# Refilter columns to be analyzed -> numerical features
numeric_features = data.select_dtypes(include=np.number)
# Create an empty list to store the percentage of outliers
results = []

# Define the threshold to define outliers
outlier_threshold = 1.5

# Initialize dictionary to store the percentage of outliers
outliers_percentage = {}

# Calculate the IQR (Interquartile Range) values and outliers
for attribute in numeric_features:
    Q1 = data[attribute].quantile(0.25)
    Q3 = data[attribute].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate lower and upper bounds
    lower_bound = Q1 - outlier_threshold * IQR
    upper_bound = Q3 + outlier_threshold * IQR

    # Identify outliers
    outliers_lower = data[data[attribute] < lower_bound]
    outliers_upper = data[data[attribute] > upper_bound]

    # Calculate percentage of lower bound outliers
    percentage_lower = len(outliers_lower) / len(data) * 100

    # Calculate percentage of upper bound outliers
    percentage_upper = len(outliers_upper) / len(data) * 100

    # Append results to the list
    results.append([attribute, f"{percentage_lower:.2f}%", f"{percentage_upper:.2f}%"])

# Create a table using PrettyTable
table = PrettyTable(["Attribute", "Lower Bound", "Upper Bound"])

# Insert data into the "Outliers Percentage" table
for result in results:
    table.add_row(result)

# Set column alignment
table.align["Attribute"] = "l"
table.align["Lower Bound"] = "c"
table.align["Upper Bound"] = "c"

# Print the table
print("Outliers Percentage:")
print(table)
```

**Insights :**

The table illustrates the percentage of outliers detected in various attributes:

- **Facebook, GooglePlus, LinkedIn:** Outliers were observed in approximately 0.00% to 16.00% of the data for these social media popularity indices.
- **SentimentTitle, SentimentHeadline:** Outliers accounted for around 1.07% to 2.58% of the sentiments scores.
- **PublishYear, PublishMonth, PublishDay, PublishHour:** Negligible outliers were found in these date-related attributes.
- **Headline_WordCount:** Around 1.45% to 12.76% of the headline word count data displayed outliers.
- **Title_WordCount:** A lower percentage of outliers, approximately 0.70% to 0.91%, was observed in the word count of the article titles.

Based on the data above, there are several significantly high outliers in the dataset. This occurs due to natural factors caused by the length of the title and headline. As the length of the headline, title, and news popularity are considered crucial factors in this research, there will be no further handling to assess the impact of this data.

**4.8.1 Hypothesis Test**

*Conducting a hypothesis test with an Independent t-test to examine if there is a significant difference in the popularity of news between articles with positive sentiment and articles with negative sentiment.*

H0: There is no significant difference between the two groups (positive sentiment and negative sentiment) regarding the popularity of news.

H1: There is a significant difference between the two groups regarding the popularity of news.

```python
# Dividing the data into two groups based on positive and negative sentiment in the title
title_positive = data[data['SentimentTitle'] > 0]
title_negative = data[data['SentimentTitle'] < 0]

# Performing Mann-Whitney U test between two groups
u_statistic, mw_p_value = mannwhitneyu(title_positive['Facebook'], title_negative['Facebook'])

# Displaying the results of the Mann-Whitney U test
if mw_p_value < 0.05:
    print("The Mann-Whitney U test indicates a significant difference between the two groups in terms of news popularity.")
else:
    print("The Mann-Whitney U test does not show a significant difference between the two groups in terms of news popularity.")
```

**Insight :**

The Mann-Whitney U test reveals a significant difference in news popularity in Facebook between articles with positive and negative sentiments. This implies that sentiment strongly influences how news articles perform in terms of engagement, specifically on Facebook. Crafting titles with positive sentiments might enhance article popularity on social media platforms. This insight can guide content strategies and marketing efforts to better engage audiences.

```python
# Performing Mann-Whitney U test between two groups
u_statistic, mw_p_value = mannwhitneyu(title_positive['LinkedIn'], title_negative['LinkedIn'])

# Displaying the results of the Mann-Whitney U test
if mw_p_value < 0.05:
    print("The Mann-Whitney U test indicates a significant difference between the two groups in terms of news popularity.")
else:
    print("The Mann-Whitney U test does not show a significant difference between the two groups in terms of news popularity.")
```

**Insight :**

Mann-Whitney U test does not yield a significant difference between the two groups concerning news popularity in LinkedIn, it suggests that the sentiment within headlines might not be a strong determinant of the article's popularity on LinkedIn. Other factors beyond sentiment might play a more critical role in influencing the engagement level of news articles on this platform. Further investigation into these additional factors, such as content relevance, timing of publication, or audience targeting, could provide deeper insights into enhancing news article performance on LinkedIn.

```python
# Performing Mann-Whitney U test between two groups
u_statistic, mw_p_value = mannwhitneyu(title_positive['GooglePlus'], title_negative['GooglePlus'])

# Displaying the results of the Mann-Whitney U test
if mw_p_value < 0.05:
    print("The Mann-Whitney U test indicates a significant difference between the two groups in terms of news popularity.")
else:
    print("The Mann-Whitney U test does not show a significant difference between the two groups in terms of news popularity.")
```

**Insights :**

The significant difference indicated by the Mann-Whitney U test in terms of news popularity concerning headline sentiments on GooglePlus suggests that there is a notable impact of the sentiment within headlines on the article's popularity on this platform. This outcome implies that the sentiment conveyed in headlines plays a discernible role in influencing the engagement level and reach of news articles on GooglePlus. Optimizing the sentiment expressed in headlines could potentially contribute positively to increasing the popularity and visibility of news articles on this platform. Further analysis can delve deeper into understanding how specific sentiments affect the engagement metrics and overall performance of articles on GooglePlus.

**Conclusions of Hypothesis Testing of Headline Sentiments Towards News Popularity :**

The analysis suggests that headline sentiments have a varying impact on news popularity across different social media platforms. While sentiment might not significantly influence news popularity on LinkedIn, it notably affects the visibility and engagement of news articles on Facebook and GooglePlus.

**4.8.2 Correlation Test**

*In this section, an analysis of the correlation between features in the dataset will be conducted using Kendall Correlations. Kendall correlation is preferred for this dataset due to its robustness to outliers.*

```python
# Calculating the kendall correlations values of headline wordcount and news popularity in various social media platforms
correlation_facebook_kendall = data['Headline_WordCount'].corr(data['Facebook'], method='kendall')
correlation_googleplus_kendall = data['Headline_WordCount'].corr(data['GooglePlus'], method='kendall')
correlation_linkedin_kendall = data['Headline_WordCount'].corr(data['LinkedIn'], method='kendall')

# Displaying the results of the kendall correlations analysis
print("Kendall correlation between headline word count and Facebook popularity:", correlation_facebook_kendall)
print("Kendall correlation between headline word count and GooglePlus popularity:", correlation_googleplus_kendall)
print("Kendall correlation between headline word count and LinkedIn popularity:", correlation_linkedin_kendall)
```

**Insight :**

The Kendall correlations between headline word count and popularity on various platforms (Facebook, GooglePlus, LinkedIn) are all negative. 

These values indicate a weak negative correlation between the length of the headlines (measured by word count) and the popularity on these platforms. In simpler terms, it suggests that as the word count in headlines increases, there is a slight tendency for the popularity on these platforms to decrease slightly, albeit weakly. 

However, it's important to note that these correlations are relatively low in magnitude, indicating a fairly weak relationship between headline length and popularity on these social media platforms. Other factors might play a more significant role in determining the popularity of articles on these platforms.

**5.9 Visualization of the Comparison of Best Publishing Times on Various Social Media Platforms**

```python
# Grouping by 'PublishHour' and calculating the mean popularity for each platform
average_popularity_by_hour = data.groupby('PublishHour')[['Facebook', 'GooglePlus', 'LinkedIn']].mean()

# Plotting average popularity for Facebook in blue color
plt.figure(figsize=(10, 6))
plt.plot(average_popularity_by_hour.index, average_popularity_by_hour['Facebook'], label='Facebook', marker='o', color='blue')

# Plotting average popularity for GooglePlus in red color
plt.plot(average_popularity_by_hour.index, average_popularity_by_hour['GooglePlus'], label='GooglePlus', marker='o', color='red')

# Plotting average popularity for LinkedIn in purple color
plt.plot(average_popularity_by_hour.index, average_popularity_by_hour['LinkedIn'], label='LinkedIn', marker='o', color='purple')

# Adding title and axis labels
plt.title('Average News Popularity by Hour')
plt.xlabel('Hour (PublishHour)')
plt.ylabel('Average Popularity')

# Adding legend
plt.legend()

# Displaying the plot
plt.grid(True)
plt.show()
```

**Insights:**

From the analysis, we can observe the optimal publishing times for news across various social media platforms and their average popularity:

- Facebook -> 5:00 PM -> average popularity approximately 195.87
- GooglePlus -> 6:00 PM -> average popularity approximately 6.54
- LinkedIn -> 11:00 AM -> average popularity approximately 19.71

Based on these findings, it can be concluded that scheduling news publications at optimal times significantly impacts achieving better popularity across social media platforms. However, to attain higher engagement, it's crucial to consider the target audience and the type of news published, as audience preferences might vary depending on the platform and the content presented.
