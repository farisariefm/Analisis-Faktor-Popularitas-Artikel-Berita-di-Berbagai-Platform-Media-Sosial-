    -> next step : Analisis Statistik Deskriptif

**Analisis Statistik Deskriptif:**

    - Menghitung statistik deskriptif seperti rata-rata, median, dan standard deviasi dari panjang headline artikel. -> Identifikasi outlier dalam data jika ada.

```python
data.loc[:, 'WordCount'] = data['Headline'].str.split().str.len()

#menginisiasi suatu kolom baru yang berisi jumlah kata pada headline artikel berita
```

```python
#melakukan perhitungan statistik dekriptif (mean, median, dan standar deviasi)
mean_wordcount = data['WordCount'].mean()
median_wordcount = data['WordCount'].median()
std_dev_wordcount = data['WordCount'].std()

#mencetak hasil analisis statistik deskriptif
print(f"Rata-rata jumlah kata pada headline artikel berita: {mean_wordcount:.2f}")
print(f"Median jumlah kata pada headline artikel berita: {median_wordcount}")
print(f"Standard Deviasi jumlah kata pada headline artikel berita: {std_dev_wordcount:.2f}")
```

Insight yang didapat :

Rata-rata jumlah kata pada headline artikel berita yang terdapat dalam dataset ini adalah 26.55 kata


Median/Nilai Tengah dari jumlah kata pada headline artikel berita yang terdapat dalam dataset ini adalah 23 kata -> rata-rata > nilai tengah -> perlu dilihat distribusi datanya 

Standard deviasi dari jumlah kata pada headline artikel berita yang terdapat di dataset ini adalah 11.79, yang dapat tergolong sebagai relatif rendah -> sebagian besar artikel memiliki jumlah kata yang cukup mendekati rata-rata, yaitu pada rentang 14.76 (dari 26.55-11.79) hingga 38.34 (26.55+11.79).

-> Distribusi jumlah kata pada headline artikel berita ini tergolong relatif konsisten, dengan mayoritas artikel berada dalam kisaran rentang tersebut. Namun, kemungkinan outlier masih ada.

-> Next : Mencari tahu jumlah outliers pada data


```python
#menghitung Banyaknya Outliers :
 
Q1 = data['WordCount'].quantile(0.25)
Q3 = data['WordCount'].quantile(0.75)
IQR = Q3 - Q1
Upper_Bound = Q3 + 1.5 * IQR
Lower_Bound = Q1 - 1.5 * IQR
#menetapkan batas IQR Range

outliers_upper = len(data[data['WordCount'] > Upper_Bound])
outliers_lower = len(data[data['WordCount'] < Lower_Bound])
#menentukan kriteria dari upper bound outliers dan lower bound outliers

total_outliers = outliers_upper + outliers_lower
#menghitung jumlah outliers (upper and lower bound)
total_data = len(data)
#menghitung jumlah entry data yang dimiliki

percentage_outliers = (total_outliers / total_data) * 100
#menghitung persentase outliers yang ada dalam data

#menampilkan hasil perhitungan outliers
print("Jumlah Outlier (Upper Bound):", outliers_upper)
print("Jumlah Outlier (Lower Bound):", outliers_lower)
print("Jumlah Total Outlier:", total_outliers)
print("Total Data:", total_data)
print("Persentase Outlier:", percentage_outliers, "%")   
```

Dari hasil analisis tersebut, berhasil didapatkan data seputar outliers dalam data ini, yaitu :

**Jumlah Outliers** pada data ini adalah 7.925 outliers yang terbagi menjadi 810 lower bound outliers, dan 7.115 upper bound outliers.

Jika dibandingkan dengan jumlah data yang dimiliki (55.757 entries), maka jumlah outliers (7.925) memiliki persentase 14.21 % yang tergolong tidak signifikan karena sebagian besar data yang ada memiliki panjang yang relatif dekat dengan nilai median ataupun rata-rata dan masih termasuk ke dalam IQR Range (Q1 hingga Q3).

**-> Next Step : Analisis Statistik Inferensial**

**Analisis Statistik Inferensial:**

    - Melakukan uji hipotesis untuk memeriksa apakah ada perbedaan signifikan dalam popularitas berita antara artikel dengan sentimen judul yang positif dan negatif.

    - Menghitung korelasi dengan metode pearson untuk menentukan sejauh mana hubungan antara panjang headline artikel dan popularitas berita pada platform-platform media sosial.

    - Menghitung korelasi hubungan antara sentimen judul dan headline dengan popularitas berita di berbagai platform media sosial.

**Melakukan uji hipotesis dengan Independent t-test untuk memeriksa apakah ada perbedaan signifikan dalam popularitas berita antara artikel dengan sentimen positif dan sentimen negatif.**

H0 : Tidak ada perbedaan signifikan antara kedua kelompok (sentimen positif dan sentimen negatif) dalam hal popularitas berita

H1 : Terdapat perbedaan signifikan antara kedua kelompok dalam hal popularitas berita


```python
title_positif = data[data['SentimentTitle']>0]
title_negatif = data[data['SentimentTitle']<0]
#membagi data menjadi dua kelompok, berdasarkan sentiment title yang positif dan negatif

t_stat,p_value = ttest_ind(title_positif['Facebook'], title_negatif['Facebook'], equal_var=False)
#membandingkan popularitas berita di Facebook antara artikel dengan sentimen positif dan sentimen negatif.

#menampilkan kondisional dari hasil uji hipotesis 

if p_value < 0.05:
    print ("Hipotesis Nol (H0) ditolak, terdapat perbedaan signifikan antara kedua kelompok terhadap popularitas berita")

else :
    print ("Hipotesis Nol (H0) diterima, tidak terdapat perbedaan signifikan antara kedua kelompok terhadap popularitas berita")
```

-> Popularitas berita di Facebook tidak dipengaruhi secara signifikan oleh sentimen dari judul artikel.

-> Faktor-faktor lain, seperti konten artikel, waktu publikasi, atau gaya penulisan, nampaknya lebih memengaruhi popularitas suatu berita di Facebook daripada sentimen judul artikel.

    -> Diterimanya H0 pada uji hipotesis ini menunjukkan bahwa popularitas berita di Facebook cenderung tidak terpengaruh oleh apakah judul artikel memiliki sentimen positif atau negatif. 

```python
title_positif = data[data['SentimentTitle']>0]
title_negatif = data[data['SentimentTitle']<0]
#membagi data menjadi dua kelompok, berdasarkan sentiment title yang positif dan negatif

t_stat,p_value = ttest_ind(title_positif['LinkedIn'], title_negatif['LinkedIn'], equal_var=False)
#membandingkan popularitas berita di LinkedIn antara artikel dengan sentimen positif dan sentimen negatif.

#menampilkan kondisional dari hasil uji hipotesis 

if p_value < 0.05:
    print ("Hipotesis Nol (H0) ditolak, terdapat perbedaan signifikan antara kedua kelompok terhadap popularitas berita")

else :
    print ("Hipotesis Nol (H0) diterima, tidak terdapat perbedaan signifikan antara kedua kelompok terhadap popularitas berita")
```

Hasil yang serupa dengan uji hipotesis sebelumnya juga didapatkan pada uji hipotesis membandingkan sentimen judul artikel terhadap popularitas berita di LinkedIn.

-> Tidak terdapat perbedaan signifikan antara kedua kelompok terhadap popularitas berita

```python
title_positif = data[data['SentimentTitle']>0]
title_negatif = data[data['SentimentTitle']<0]
#membagi data menjadi dua kelompok, berdasarkan sentiment title yang positif dan negatif

t_stat,p_value = ttest_ind(title_positif['GooglePlus'], title_negatif['GooglePlus'], equal_var=False)
#membandingkan popularitas berita di GooglePlus antara artikel dengan sentimen positif dan sentimen negatif.

#menampilkan kondisional dari hasil uji hipotesis 

if p_value < 0.05:
    print ("Hipotesis Nol (H0) ditolak, terdapat perbedaan signifikan antara kedua kelompok terhadap popularitas berita")

else :
    print ("Hipotesis Nol (H0) diterima, tidak terdapat perbedaan signifikan antara kedua kelompok terhadap popularitas berita")
    
print(p_value)
```

**Insights :**

Ditolaknya H0 pada perbandingan antara sentimen positif dan sentimen negatif pada popularitas artikel berita di platform GooglePlus menunjukkan bahwa :

Untuk Facebook dan LinkedIn, tidak terdapat perbedaan signifikan dalam popularitas berita antara artikel dengan sentimen positif dan negatif. Dengan kata lain, popularitas berita di kedua platform tersebut tidak dipengaruhi oleh sentimen artikel.

Namun, di platform GooglePlus, terdapat perbedaan signifikan dalam popularitas berita antara artikel dengan sentimen positif dan sentimen negatif. Hasil ini juga menunjukkan bahwa adanya kemungkinan pengaruh sentimen judul artikel terhadap popularitas beritanya -> diperlukan analisis lebih lanjut (analisis korelasi sentiment title dan headline terhadap popularitas berita).

-> Next Step : Menghitung korelasi untuk menentukan sejauh mana hubungan antara panjang headline artikel dan popularitas berita pada platform-platform media sosial.

**Menghitung korelasi untuk menentukan sejauh mana hubungan antara panjang headline artikel dan popularitas berita pada platform-platform media sosial.**

```python
#menghitung korelasi antara jumlah kata pada headline artikel dengan popularitas berita di platform Facebook
correlation_facebook = data['WordCount'].corr(data['Facebook'])

#menghitung korelasi antara jumlah kata pada headline artikel dengan popularitas berita di platform GooglePlus
correlation_googleplus = data['WordCount'].corr(data['GooglePlus'])

#menghitung korelasi antara jumlah kata pada headline artikel dengan popularitas berita di platform LinkedIn
correlation_linkedin = data['WordCount'].corr(data['LinkedIn'])

#menampilkan hasil korelasi
print("Korelasi antara jumlah kata dan popularitas di Facebook:", correlation_facebook)
print("Korelasi antara jumlah kata dan popularitas di GooglePlus:", correlation_googleplus)
print("Korelasi antara jumlah kata dan popularitas di LinkedIn:", correlation_linkedin)
```

**Insights :**

Korelasi antara jumlah kata pada headline artikel dengan popularitas berita di Facebook adalah -0.0501. 

    -> Ini menunjukkan bahwa terdapat korelasi negatif yang sangat lemah antara jumlah kata dalam headline artikel dan popularitas di Facebook. Artinya, semakin panjang artikel, popularitas di Facebook cenderung sedikit menurun, tetapi korelasinya sangat lemah.

Korelasi antara jumlah kata pada headline artikel dan popularitas berita di GooglePlus adalah -0.0554. 

    -> Sama seperti di Facebook, terdapat korelasi negatif yang sangat lemah antara jumlah kata dalam headline artikel dan popularitas di GooglePlus. Artinya, semakin panjang artikel, popularitas di GooglePlus cenderung sedikit menurun, tetapi korelasinya sangat lemah.

Korelasi antara jumlah kata pada headline artikel dan popularitas berita di LinkedIn adalah -0.0388. 

    -> Korelasi negatif juga terdapat di LinkedIn, tetapi korelasinya lebih lemah dibandingkan dengan Facebook dan GooglePlus. Ini berarti jumlah kata dalam headline artikel memiliki pengaruh yang sangat lemah terhadap popularitas di LinkedIn.

Korelasi negatif yang lemah menunjukkan bahwa jumlah kata pada headline artikel hanya memiliki pengaruh yang sangat kecil terhadap popularitas berita di berbagai platform media sosial tersebut. Jadi, secara umum, panjang (jumlah kata) headline artikel tidak secara signifikan memengaruhi popularitas berita di ketiga platform tersebut.


-> Next : Analisa Korelasi Sentiment Title dan Headline Artikel Berita terhadap Popularitas Berita di Berbagai Platform Media Sosial.

```python
#menghitung korelasi antara SentimentTitle, SentimentHeadline, dan popularitas di setiap platform
correlation_facebook_title = data['SentimentTitle'].corr(data['Facebook'])
correlation_facebook_headline = data['SentimentHeadline'].corr(data['Facebook'])

correlation_googleplus_title = data['SentimentTitle'].corr(data['GooglePlus'])
correlation_googleplus_headline = data['SentimentHeadline'].corr(data['GooglePlus'])

correlation_linkedin_title = data['SentimentTitle'].corr(data['LinkedIn'])
correlation_linkedin_headline = data['SentimentHeadline'].corr(data['LinkedIn'])

#menampilkan hasil analisis
print("Korelasi SentimentTitle dengan Facebook:", correlation_facebook_title)
print("Korelasi SentimentHeadline dengan Facebook:", correlation_facebook_headline)
print("Korelasi SentimentTitle dengan GooglePlus:", correlation_googleplus_title)
print("Korelasi SentimentHeadline dengan GooglePlus:", correlation_googleplus_headline)
print("Korelasi SentimentTitle dengan LinkedIn:", correlation_linkedin_title)
print("Korelasi SentimentHeadline dengan LinkedIn:", correlation_linkedin_headline)
```

**Insight yang Didapat :**

Korelasi Sentiment Title dan Headline dengan popularitas berita di platform Facebook menunjukkan bahwa tidak ada hubungan yang kuat antara sentimen judul dan headline sebuah artikel dengan popularitas berita di platform Facebook -> Korelasi ini mendekati nol, yang berarti sentimen judul dan headline artikel tidak mempengaruhi popularitas di Facebook.

Untuk platform GooglePlus, korelasi antara Sentiment Title dengan popularitas beritanya adalah negatif, namun belum cukup kuat. Korelasi Sentiment Headline dengan popularitas berita di GooglePlus bersifat positif, tetapi masih lemah, dan tidak begitu kuat seperti Sentiment Title terhadap popularitas beritanya.

Di platform LinkedIn, korelasi antara Sentiment Title dan Headline terhadap popularitas beritanya juga tergolong tidak kuat.

Dalam beberapa platform, terdapat korelasi positif antara sentimen judul dan headline dengan popularitas, meskipun tidak sangat kuat. Namun, perlu diingat bahwa faktor-faktor lain seperti waktu publikasi dan topik berita juga dapat memengaruhi popularitas di masing-masing platform.
